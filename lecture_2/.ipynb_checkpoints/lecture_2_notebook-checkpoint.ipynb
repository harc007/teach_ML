# Reading using pandas

```python
#Load from string
data = 'col1,col2,col3\na,b,1\na,b,2\nc,d,3'

df = pd.DataFrame([row.split(',') for row in data.split('\n')])

df.columns = df.iloc[0]

df.reindex(df.index.drop(0))
```
