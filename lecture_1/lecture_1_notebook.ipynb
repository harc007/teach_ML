IMPORTING FOLLOWING PACKAGES:
1. Numpy - statistical computation
2. stats from scipy package - Generating distributions
3. pyplot from matplotlib package - Visualizations

```python
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
```

# BINOMIAL DISTRIBUTION

```python
n = 10
p = 0.3
k = np.arange(0, 21)
k
```

```python
binomial = stats.binom.pmf(k, n, p)
binomial
```

```python
plt.plot(k, binomial, 'o-')
plt.title('Binomial: n=%i, p=%.2f' % (n,p), fontsize=15)
plt.xlabel('Number of Successes')
plt.ylabel('Probability of successes', fontsize=15)
plt.show()
```

```python
binom_sim = data = stats.binom.rvs(n=10, p=0.3, size=10000)
print "Mean: %g" % np.mean(binom_sim)
print "SD: %g" % np.std(binom_sim, ddof=1)
plt.hist(binom_sim, bins=10, normed=True)
plt.xlabel("X")
plt.ylabel("Density")
plt.show()
```

# POISSON DISTRIBUTION

```python
rate = 2
n = np.arange(0, 10)
y = stats.poisson.pmf(n, rate)
y
```

```python
plt.plot(n, y, 'o-')
plt.title('Poisson: $\lambda$ = %i' % rate)
plt.xlabel('Number of accidents')
plt.ylabel('Probability of number of accidents')
plt.show()
```

```python
data = stats.poisson.rvs(mu=2, loc=0, size=1000)
print "Mean: %g" % np.mean(data)
print "SD: %g" % np.std(data, ddof=1)
plt.figure()
plt.hist(data, bins=9, normed=True)
plt.xlim(0, 10)
plt.xlabel("Number of accidents")
plt.title("Simulating Poisson random variables")
plt.show()
```

# NORMAL DISTRIBUTION

```python
mu = 0
sigma = 1
x = np.arange(-5, 5, 0.1)
y = stats.norm.pdf(x, 0, 1)
plt.plot(x, y)
plt.title('Normal: $\mu$= %.1f $\sigma^2$=%.1f' % (mu, sigma))
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.show()
```

```python
data = stats.norm.rvs(size=1000)
plt.hist(data, bins=20, normed=True)
plt.xlim(-5, 5)
plt.xlabel("Deviation from average height")
plt.title("Simulating Normal random variables")
plt.show()
```
