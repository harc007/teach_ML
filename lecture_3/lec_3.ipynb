```python
df = pd.read_csv('adult.csv', header=None)
df.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 
             'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 
             'class']
df.head(5)
```

# Create new features / Feature Engineering

```python
df["difficulty"] = df["age"] * df["hours-per-week"]
df.head(5)
```

```python
# Functional programming pandas
df["gender_num"] = df["sex"].apply(lambda x: 1 if x == ' Female' else 0)
df[["age", "sex", "gender_num"]].head(5)
```

```python
df["race"].unique()
def assign_race_number(x):
    if x == ' White':
        return 0
    elif x == ' Black':
        return 1
    elif x == ' Asian-Pac-Islander':
        return 2
    elif x == ' Amer-Indian-Eskimo':
        return 3
    else:
        return 4

df["race_num"] = df["race"].apply(lambda x: assign_race_number(x))
df[["race", "race_num"]].head(5)
```

# Merge and Joining

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4'], 
                     'A': ['A0', 'A1', 'A2', 'A3', 'A4'], 
                     'B': ['B0', 'B1', 'B2', 'B3', 'B4']})
left
```

```python
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K5'], 
                      'C': ['C0', 'C1', 'C2', 'C3', 'C5'], 
                      'D': ['D0', 'D1', 'D2', 'D3', 'D5']})
right
```

```python
# Left join
pd.merge(left, right, on='key', how='left')
```

```python
# Right join
pd.merge(left, right, # Left join
pd.merge(left, right, on='key', how='left')on='key', how='right')
```

```python
# Outer join
pd.merge(left, right, on='key', how='outer')
```

```python
# Inner join
pd.merge(left, right, on='key', how='inner')
```

```python
df = pd.concat([left, right])
df.index = range(df.shape[0])
df
```
